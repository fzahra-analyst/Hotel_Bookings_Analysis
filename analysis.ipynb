```python
df['country'].fillna('Unknown', inplace=True)
```

```python
cancel_rate = df['is_canceled'].mean() * 100
print(f"معدل الإلغاء العام هو: {cancel_rate:.2f}%")
```

```python
cancel_by_hotel = df.groupby('hotel')['is_canceled'].mean() * 100
print(cancel_by_hotel)
```

```python
sns.barplot(x='hotel', y='is_canceled', data=df)
plt.title("Cancellation rate by hotel type")
plt.ylabel("Cancellation rate (%)")
plt.xlabel("Hotel type")
plt.show()
```

```python
df['cancel_rate'] = df['hotel'].map(cancel_by_hotel)
```

```python
print(df[['hotel', 'cancel_rate']].head())
```

```python
sns.barplot(x='hotel', y='cancel_rate', data=df)
plt.title("Cancellation rate by hotel type")
plt.ylabel("Cancellation rate")
plt.xlabel("Hotel type")
plt.show()
```

```python
sns.boxplot(x='is_canceled', y='lead_time', data=df)
plt.title("The relationship between advance booking duration and cancellation")
plt.ylabel("Number of days between booking and arrival")
plt.xlabel("is canceled")
plt.show()
```

```python
print(df['adr'].describe())
```

```python
# إزالة القيم التي فيها adr أكبر من 1000 مثلاً
df_clean = df[df['adr'] < 1000]

sns.boxplot(x='is_canceled', y='adr', data=df_clean)
plt.title("The relationship between the daily rate and the likelihood of cancellation")
plt.ylabel("Daily price (ADR)")
plt.xlabel("is canceled")
plt.show()
```

```python
sns.barplot(x='deposit_type', y='is_canceled', data=df)
plt.title("Cancellation rate by deposit type")
plt.ylabel("Cancellation rate (%)")
plt.xlabel("Deposit type")
plt.show()
```

```python
top_countries = df['country'].value_counts().head(10).index
sns.barplot(x='country', y='is_canceled', data=df[df['country'].isin(top_countries)])
plt.title("Cancellation rate for the top 10 booking countries")
plt.ylabel("Cancellation rate (%)")
plt.xlabel("country")
plt.show()
```
